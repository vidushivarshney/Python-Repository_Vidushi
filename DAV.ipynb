```python
cat_data=data['Genre'].value_counts()

x_bar=cat_data.index
y_bar=cat_data

plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(16,6))
plt.xlabel("genre",loc="left")
plt.ylabel("count")
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=15)
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(16,6))
plt.xlabel("genre",fontsize=20)
plt.ylabel("count",fontsize=20)
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=15)
plt.title("Count by Genre")

plt.bar(x_bar,y_bar ,color="orange",width=0.2 )
```
